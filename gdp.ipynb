```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
years = 2012 - 1990 + 1
gdp_year = np.zeros(years) 

tmp_avg_year = np.zeros(years, dtype=float)
tmp_var_year = np.zeros(years, dtype=float)
```

```python
gdpfile = open("temperature_GDP_data.csv")
lines = gdpfile.readlines()
for i in range(years):
    tmp_avg = 0
    tmp_sum = 0
    gdp = 0
    splited = ''
    tmp_month= np.zeros(12) 
    
    for j in range(12):
        line = lines[i*12+j]
        splited = line.split(',')
        tmp_month[j] = float(splited[1])
        assert(int(splited[0][:4])==i+1960)
        tmp_sum += float(splited[1])
    gdp = float(splited[4])
    assert(gdp>0.001)
    tmp_avg = tmp_month.var()
    gdp_year[i] = gdp
    tmp_avg_year[i] = tmp_avg
    tmp_var_year[i] = tmp_month.var()
```

```python
from sklearn import linear_model
```

```python
years = 2013 - 1960 + 1

tmp_avg_year = np.zeros(years, dtype=float)
tmp_var_year = np.zeros(years, dtype=float)

gdpfile = open("year_temperature.csv")
lines = gdpfile.readlines()
for i in range(years):
    splited = ''
    line = lines[i]
    splited = line.split(',')

    tmp_avg = splited[1]

    tmp_avg_year[i] = tmp_avg
```

```python
year_x = [i for i in range(1960, 2013+1)]
```

```python
tmp_avg_year[0:3]
```

```python
plt.title("Temperature") 
plt.xlabel("Years") 
plt.ylabel("Temperature") 
# plt.plot(year_x,gdp_year) 
plt.plot(year_x,tmp_avg_year) 
plt.show()
```

```python
win_size = 10
train_years = years -20
train_x = np.zeros((train_years-win_size, win_size + 2), dtype=float)
train_y = np.zeros(train_years-win_size, dtype=float)

for i in range(train_years - win_size):
    for j in range(win_size):
        train_x[i][j] = tmp_avg_year[i+j]
    train_x[i][win_size] = np.max(tmp_avg_year[i:i+win_size])
    train_x[i][win_size+1] = np.min(tmp_avg_year[i:i+win_size])
    train_y[i] = tmp_avg_year[i+win_size] + 1
    
    
```

```python
model = linear_model.LinearRegression()
model.fit(train_x, train_y)
```

```python
display(model.intercept_)
display(model.coef_)
```

```python
tmp_test = tmp_avg_year.copy()
tmp_test= np.append(tmp_test,np.zeros(50, dtype=float))

test_y = np.zeros(50, dtype=float)
for i in range(years-win_size, years+50-win_size):
    test_x = np.zeros(win_size+2, dtype=float)
    for j in range(win_size):
        test_x[j] = tmp_test[i+j]
    test_x[win_size] = np.max(tmp_test[i:i+win_size])
    test_x[win_size+1] = np.min(tmp_test[i:i+win_size])
    test_y[i-years] = model.predict([test_x])
    tmp_test[i+win_size] = test_y[i-years]
```
